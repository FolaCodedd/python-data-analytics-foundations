**Part 2**: Vectors using pandas and NumPy
1. Select one numeric column as a pandas Series.
2. Convert the column to a NumPy array.
3. Calculate the mean, minimum, and maximum.
4. Increase all values by 10% using vectorized operations.

```python
import numpy as np

Sales_Vector = df['Sales'].to_numpy()

Calculations = ['Mean','Minimum','Maximum']

for calc in Calculations:
    match calc:
        case 'Mean': result = Sales_Vector.mean()
        case 'Minimum': result = Sales_Vector.min()
        case 'Maximum': result = Sales_Vector.max()

    print(f"The {calc} sales is {result}")

print('___________________________________________________________________________________________________________________')

# create vectorized function
increase_by_10_percent = lambda i: i * 1.10

vectorized_increase = np.vectorize(increase_by_10_percent)
print(vectorized_increase(Sales_Vector))

```

**Part 3**: Arrays and Matrices
1. Select multiple numeric columns.
2. Convert them into a NumPy array (matrix).
3. Display the shape of the array.
4. Compute column-wise means and row-wise sums.

```python
Quantity_Profit = df[['Quantity','Profit']]

new_matrix = np.array(Quantity_Profit)

print(f"The shape of the matrix is: {new_matrix.shape}")

Column_Means = np.mean(new_matrix,axis=0)

Row_Sums = np.sum(new_matrix,axis=1)

print(f"\n The column-wise means are: \n {Column_Means}")
print(f"\n The row-wise sums are: \n {Row_Sums}")

```

**Part 4**: Matrix Operations
1. Normalize the matrix values.
2. Transpose the matrix.
3. Multiply the matrix by a scalar.

```python

matrix_min = new_matrix.min()
matrix_max = new_matrix.max()

# Apply Min-Max normalization
normalized_matrix = (new_matrix - matrix_min) / (matrix_max - matrix_min)

print("\nNormalized Matrix (Min-Max):")
print(normalized_matrix)

print("\nTranposed Matrix:")
print(normalized_matrix.T)

print("\n Scalar Multiplication:")
print(normalized_matrix * 10)
```
